# 🍽️ Zomato Dataset - Exploratory Data Analysis (EDA)

This project performs an exploratory data analysis on the Zomato dataset to uncover insights about restaurants, ratings, pricing, and customer preferences.

## 📌 Problem Statement

The goal of this analysis is to understand the patterns in restaurant data such as ratings, cost, location, and services like online delivery and table booking.

We aim to answer questions like:
- Which restaurants have the highest ratings?
- How does cost affect ratings?
- Which cities have the most restaurants?
- What services are most common?

## 📦 Importing Required Libraries

We import essential Python libraries for data analysis and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

## 📂 Loading the Dataset

The dataset is loaded into a Pandas DataFrame for further analysis.

```python
df=pd.read_csv('Zomato_Data.csv',encoding='latin-1')
```

## 🧹 Data Cleaning

In this step, we handle:
- Missing values
- Incorrect data types
- Duplicate entries

This ensures the dataset is clean and ready for analysis.

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
df.isna().sum()
```

```python
df=df.drop(columns=['Longitude','Rating color','Switch to order menu','Latitude','Locality Verbose'])
```

```python
df.head()
```

```python
df[df.duplicated() == True]
```

```python
df=df.reset_index(drop=True)
```

```python
df.dtypes
```

```python
df['Country Code']=df['Country Code'].astype('category')
df['City']=df['City'].astype('category')
df['Locality']=df['Locality'].astype('category')
df['Currency']=df['Currency'].astype('category')
```

```python
df.dtypes
```

```python
df['Rating text']=df['Rating text'].astype('category')
```

```python
df=df.rename(columns={'Has Table booking':'Table booking',
                        'Has Online delivery':'Online delivery',
                        'Is delivering now':'Delivering now',
                        'Aggregate rating':'Rating'
               })
```

```python
df['Table booking']=df['Table booking'].astype('bool')
df['Online delivery']=df['Online delivery'].astype('bool')
df['Delivering now']=df['Delivering now'].astype('bool')
```

```python
df.head()
```

## 📊 Exploratory Data Analysis (EDA)
We explore the dataset to understand patterns, trends, and relationships between variables.

### ⭐ Top 10 Restaurants by Rating
Displays the highest rated restaurants in the dataset.

```python
top_rest = df.sort_values(by='Rating', ascending=False).head(10)

plt.figure()
sns.barplot(data=top_rest, x='Rating', y='Restaurant Name', palette='viridis')

plt.title('Top 10 Restaurants by Rating')
plt.xlabel('Rating')
plt.ylabel('Restaurant Name')
plt.show()
```

### 🍽️ Top 10 Cuisines
Shows the most popular cuisines based on restaurant frequency in the dataset.

```python
cuisine = df['Cuisines'].str.split(', ').explode()
top_cuisines = cuisine.value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index)
plt.title('Top 10 Cuisines')
plt.show()
```

### 📦 Box Plot of Restaurant Ratings
Shows the distribution and spread of restaurant ratings, including outliers.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.boxplot(y=df['Rating'],palette='husl')
plt.title('Box Plot of Restaurant Ratings')
plt.show()
```

### 💰 Price Range Distribution
Shows how restaurant price ranges are distributed in the dataset.

```python
plt.figure()
sns.countplot(data=df, x='Price range', palette='pastel')
plt.title('Price Range Distribution')
plt.xticks(rotation=90)
plt.show()
```

### 🔗 Relationship Between Variables
This pairplot shows relationships between rating, votes, cost, and price range.

```python
pair_df = df[['Rating', 'Votes',
              'Average Cost for two',
              'Price range']].dropna()
sns.pairplot(
    pair_df,
    diag_kind='kde',
    plot_kws={
        'alpha': 0.5,
        'color': '#7F77DD',
        'edgecolor': 'none'
    },
    diag_kws={
        'color': '#534AB7',
        'fill': True,
        'alpha': 0.6
    }
)
plt.suptitle('Pairplot — rating, votes, cost & price range',            fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

### 💸 Does Higher Cost Mean Better Rating?
This visualization explores whether expensive restaurants tend to have higher ratings and compares online vs offline delivery.

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df.sample(min(1500, len(df)), random_state=42),
    x='Average Cost for two',
    y='Rating',
    hue='Online delivery',
    palette={True: '#1D9E75', False: '#D85A30'},
    alpha=0.55,
    s=70,
    edgecolor='none'
)

plt.title('Does higher cost mean better rating?',
          fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Average cost for two', fontsize=13)
plt.ylabel('Rating', fontsize=13)
plt.legend(title='Online delivery', fontsize=11)
plt.tight_layout()
plt.show()
```

### ⭐ Rating Category Distribution
Shows how restaurants are distributed across different rating labels.

```python
order = df['Rating text'].value_counts().index
plt.figure(figsize=(10, 6))
sns.countplot(
    data=df,
    x='Rating text',
    order=order,
    palette='RdYlGn'
)
plt.title('Count of restaurants by rating category',
          fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Rating text', fontsize=13)
plt.ylabel('Count', fontsize=13)
plt.xticks(rotation=30, ha='right', fontsize=11)
plt.tight_layout()
plt.show()
```

### 📊 Votes by Rating Category
Shows how user votes are distributed across different rating groups.

```python
plt.figure(figsize=(10,6))
order = ['Excellent', 'Very Good', 'Good', 'Average', 'Poor']
order = [o for o in order if o in df['Rating text'].unique()]
sns.violinplot(
    data=df[df['Rating text'].isin(order)],
    x='Rating text',
    y='Votes',
    order=order,
    palette=['#1D9E75', '#378ADD', '#EF9F27', '#F0997B', '#E24B4A'],
    inner='box',
    linewidth=1.5,
    cut=0
)
plt.title('Votes distribution by rating category',
          fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Rating category', fontsize=13)
plt.ylabel('Votes', fontsize=13)
plt.tight_layout()
plt.show()
```

## 📊 Overview Dashboard (EDA Summary)
A combined view of key insights including ratings, cost, votes, cuisines, price range, and correlations.

```python
sns.set_style("whitegrid")
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

sns.boxplot(y=df['Rating'], color='orange', ax=axes[0,0])
axes[0,0].set_title('Rating Distribution (Boxplot)')

cuisine = df['Cuisines'].str.split(', ').explode()
top_cuisines = cuisine.value_counts().head(5)

sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette='viridis', ax=axes[0,1])
axes[0,1].set_title('Top Cuisines')

sns.scatterplot(data=df, x='Average Cost for two', y='Rating', color='red', ax=axes[0,2])
axes[0,2].set_title('Cost vs Rating')

sns.scatterplot(data=df, x='Votes', y='Rating', color='blue', ax=axes[1,0])
axes[1,0].set_title('Votes vs Rating')

sns.countplot(data=df, x='Price range', palette='Set2', ax=axes[1,1])
axes[1,1].set_title('Price Range Distribution')

num_df = df[['Average Cost for two', 'Rating', 'Votes']].dropna()
corr = num_df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', ax=axes[1,2])
axes[1,2].set_title('Correlation Heatmap')

plt.tight_layout()
plt.show()
```

## 📌 Final Insights
Key observations from the Zomato dataset analysis:
- Most restaurants fall under "Good" and "Average" rating categories.
- Higher cost does not always guarantee higher ratings.
- Votes and ratings show some relationship but not very strong correlation.
- North Indian and other popular cuisines dominate the dataset.
- Restaurants with online delivery tend to get higher engagement (votes).
